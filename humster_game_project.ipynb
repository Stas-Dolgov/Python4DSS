```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk
import time

# Глобальные переменные
coins = 0
rate = 1  # Количество монет за клик
boost_active = False
boost_rate = 5  # Увеличенное количество монет за клик во время boost
promotion_active = False  # Флаг для акции

# Функция для обновления изображения на кнопке в зависимости от количества монет
def update_button_image():
    global coins
    if coins <= 120:
        click_button.config(image=button_photo_low)
    elif coins <= 250:
        click_button.config(image=button_photo_avg)
    else:
        click_button.config(image=button_photo_high)

# Функция для увеличения количества монет при клике
def click():
    global coins, rate, boost_rate
    if boost_active:
        coins += boost_rate * rate
    else:
        coins += rate
    label.config(text=f"Coins: {coins}")
    update_button_image()
    check_coins_120()  # Проверка монет
    check_coins_250()  # Проверка монет

# Функция для проверки монет 120
def check_coins_120():
    global boost_active, rate
    if coins >= 120 and coins < 250:
        rate = 2
        if boost_active:
            click_label.config(text=f"Coins per click: {boost_rate * rate} (Boosted)")

# Функция для проверки монет 250
def check_coins_250():
    global boost_active, rate
    if coins >= 250:
        rate = 3
        if boost_active:
            click_label.config(text=f"Coins per click: {boost_rate * rate} (Boosted)")            

# Функция для вывода монет
def cash_out():
    global coins, rate, boost_active
    if coins >= 100:  # Условие для вывода монет (например, нужно 100 монет для вывода)
        coins = 0
        label.config(text="Coins: 0")
        update_button_image()
        messagebox.showinfo("Cash Out", "You have cashed out successfully!")
        rate = 1
        click_label.config(text=f"Coins per click: {rate}")
        boost_active = False
    else:
        messagebox.showwarning("Cash Out", "Not enough coins to cash out. You need at least 100 coins.")

# Функция для активации boost
def activate_boost():
    global boost_active, boost_rate
    boost_active = True
    boost_button.config(state=tk.DISABLED)  # Делаем кнопку неактивной на время boost
    click_label.config(text=f"Coins per click: {boost_rate * rate} (Boosted)")
    root.after(10000, deactivate_boost)  # Через 10 секунд отключаем boost

# Функция для деактивации boost
def deactivate_boost():
    global boost_active
    boost_active = False
    boost_button.config(state=tk.NORMAL)  # Включаем кнопку после окончания boost
    click_label.config(text=f"Coins per click: {rate}")  # Возвращаем обычное значение монет за клик

# Функция для активации акции
def activate_promotion():
    global coins, promotion_active
    if coins >= 70 and not promotion_active:
        coins -= 70
        promotion_active = True
        label.config(text=f"Coins: {coins}")
        update_button_image()
        root.after(1000, add_coin_per_second)  # Начинаем добавлять монеты каждую секунду
    else:
        messagebox.showwarning("Promotion", "Not enough coins for the promotion.")

# Функция для добавления одной монеты каждую секунду
def add_coin_per_second():
    global coins, promotion_active
    if promotion_active:
        coins += 1
        label.config(text=f"Coins: {coins}")
        update_button_image()
        root.after(1000, add_coin_per_second)  # Продолжаем добавлять монеты каждую секунду

# Создание основного окна
root = tk.Tk()
root.title("Clicker Game")
# Увеличение размера окна
root.geometry("400x500")
#Изменения цвета
root.configure(bg = '#4855bd')

# Создание метки для отображения количества монет
label = tk.Label(root, text="Coins: 0", font=("Roboto", 16), bg='#4855bd', fg = 'white')
label.pack(pady = 5)

# Создание метки для отображения текущего значения получения монет при клике
click_label = tk.Label(root, text = "Coins per click: 1", font = ("Roboto", 14), bg = '#4855bd', fg = 'white')
click_label.pack(pady = 5)

# Загрузка и изменение размера изображений для кнопок
button_image_low = Image.open('image_buttom1.png').resize((200, 200))
button_photo_low = ImageTk.PhotoImage(button_image_low)

button_image_avg = Image.open("image_buttom2.png").resize((200, 200))
button_photo_avg = ImageTk.PhotoImage(button_image_avg)

button_image_high = Image.open("image_buttom3.png").resize((200, 200))
button_photo_high = ImageTk.PhotoImage(button_image_high)

# Создание кнопки с изображением
click_button = tk.Button(root, image = button_photo_low, command = click, bd = 0, highlightthickness = 0, bg = '#4855bd', 
                         activebackground = '#4855bd')
click_button.pack(pady = 5)

# Создание кнопки для вывода монет
cash_out_button = tk.Button(root, text = "Cash Out", font=("Roboto", 16), command = cash_out, bd = 0, highlightthickness = 0, 
                            bg = 'black', fg = '#4855bd', activebackground = 'black')
cash_out_button.pack(pady = 5)

# Создание кнопки для активации boost
boost_button = tk.Button(root, text = "Boost", font = ("Roboto", 16), command = activate_boost, bg = '#d97736')
boost_button.pack(pady = 5)

# Создание кнопки для активации акции
promotion_button = tk.Button(root, text = "Promotion", font = ("Roboto", 16), command = activate_promotion, bg = '#d97736')
promotion_button.pack(pady = 5)

# Запуск главного цикла обработки событий
root.mainloop()
```
