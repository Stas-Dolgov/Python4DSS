```python
from PIL import Image, ImageDraw
#### Функция для рисования шахматной доски

def draw_chessboard(size: int, square_size: int, colors: tuple = ("white", "black"), background_color: str = "white"):
    """
    Рисует шахматную доску с заданными параметрами.

    :param size: Размер изображения (ширина и высота).
    :param square_size: Размер одной клетки шахматной доски.
    :param colors: Кортеж с двумя цветами для клеток доски (светлый и темный).
    :param background_color: Цвет фона.
    :return: Объект PIL.Image с нарисованной шахматной доской.
    """
    board_size = size * square_size
    image = Image.new('RGB', (board_size, board_size))
    draw = ImageDraw.Draw(image)

    for y in range(size):
        for x in range(size):
            fill_color = colors[(x + y) % 2]
            top_left = (x * square_size, y * square_size)
            bottom_right = ((x + 1) * square_size, (y + 1) * square_size)
            draw.rectangle([top_left, bottom_right], fill=fill_color)

    return image

# Пример 1: стандартная черно-белая шахматная доска размером 8x8
chessboard1 = draw_chessboard(size=8, square_size=50)
chessboard1
```

```python
# Пример 2: шахматная доска размером 10x10 с оранжевым и серыми клетками
chessboard2 = draw_chessboard(size=10, square_size=40, colors=("orange", "grey"))
chessboard2
```

```python

# Пример 3: шахматная доска размером 4*4 с зелеными и желтыми клетками и серым фоном
chessboard3 = draw_chessboard(size=4, square_size=40, colors=("green", "yellow"))
chessboard3
```
