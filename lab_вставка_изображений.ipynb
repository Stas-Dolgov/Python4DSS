Нарисовать превью списка картинок

```python
from PIL import Image
```

```python
image_block = Image.open('image_2.png')
print(image_block.size)
image_block
```

```python
image_block_mini = image_block.copy()
image_block_mini.thumbnail((100, 100))
image_block_mini
```

```python
im = Image.new(size=(700, 200), mode = 'RGB')

im.paste(image_block_mini)
im.paste(image_block_mini, box=(200, 0))
im.paste(image_block_mini, box=(300, 0))
im.paste(image_block_mini, box=(200, 100))
im.paste(image_block_mini, box=(200, 200))
im.paste(image_block_mini, box=(300, 100))
im
```
