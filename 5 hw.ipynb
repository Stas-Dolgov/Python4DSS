```python
import random

def transform_string(input_string):
    words = input_string.split()
    """
    swap_case В каждом слове поменять регистр символов на противоположный
    replace_with_underscore В каждом слове заменить каждую букву на '_'
    first_last_replace В каждом слове первую и последнюю букву заменить на '_'
    replace_if_short В каждом слове заменить каждую букву на '_', если длина слова меньше 5 символов
    random_shuffle В каждом слове поменять буквы местами случайным образом
    last_upper Каждую последнюю букву в слове написать с большой буквы
    alternate_case Каждое второе слово написать большими буквами, а остальные слова маленькими
    random_words Поменять слова местами случайным образом
    sorted_by_length Сформировать список слов, отсортированных по количеству букв в каждом слове
    unique_chars Сформировать список уникальных букв"""

    swap_case = [word.swapcase() for word in words]
    replace_with_underscore = ['_' * len(word) for word in words]
    first_last_replace = [('_' + word[1:-1] + '_') if len(word) > 1 else '_' for word in words]
    replace_if_short = ['_' * len(word) if len(word) < 5 else word for word in words]
    random_shuffle = [''.join(random.sample(word, len(word))) for word in words]
    last_upper = [word[:-1] + word[-1].upper() if word else '' for word in words]
    alternate_case = [word.upper() if i % 2 else word.lower() for i, word in enumerate(words)]
    random_words = random.sample(words, len(words))
    sorted_by_length = sorted(words, key=len)
    unique_chars = list(set(''.join(words)))
    
    # 11. Сформировать список уникальных букв и их количество
    from collections import Counter
    char_count = dict(Counter(''.join(words)))

    return {
        "swap_case": swap_case,
        "replace_with_underscore": replace_with_underscore,
        "first_last_replace": first_last_replace,
        "replace_if_short": replace_if_short,
        "random_shuffle": random_shuffle,
        "last_upper": last_upper,
        "alternate_case": alternate_case,
        "random_words": random_words,
        "sorted_by_length": sorted_by_length,
        "unique_chars": unique_chars,
        "char_count": char_count
    }

# Пример использования функции
input_string = "labGenComprehension - is my homework for today"
result = transform_string(input_string)

# Печать результатов
for key, value in result.items():
    print(f"{key}: {value}")
```

```python
import os
from collections import Counter

def get_file_stats(directory):
    files_data = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            file_size = os.path.getsize(file_path)
            file_name, file_ext = os.path.splitext(file)
            files_data.append({
                'name': file,
                'size': file_size,
                'name_length': len(file_name),
                'ext_length': len(file_ext),
                'ext': file_ext,
                'path_length': len(file_path),
                'path': file_path,
                'dir': root
            })
    return files_data

def show_top_files_by_size(files_data, top_n=10, largest=True):
    sorted_files = sorted(files_data, key=lambda x: x['size'], reverse=largest)
    return sorted_files[:top_n]

def show_top_files_by_name_length(files_data, top_n=10, longest=True):
    sorted_files = sorted(files_data, key=lambda x: x['name_length'], reverse=longest)
    return sorted_files[:top_n]

def show_top_files_by_ext_length(files_data, top_n=10, longest=True):
    sorted_files = sorted(files_data, key=lambda x: x['ext_length'], reverse=longest)
    return sorted_files[:top_n]

def show_top_most_common(items, top_n=10):
    counter = Counter(items)
    total = sum(counter.values())
    top_common = counter.most_common(top_n)
    return [(item, count, count / total * 100) for item, count in top_common]

def show_top_dirs_by_file_count(files_data, top_n=10):
    dir_counter = Counter([file['dir'] for file in files_data])
    return dir_counter.most_common(top_n)

def show_top_dirs_by_subdir_count(directory, top_n=10):
    subdir_counter = Counter()
    for root, dirs, files in os.walk(directory):
        subdir_counter[root] += len(dirs)
    return subdir_counter.most_common(top_n)

def main(directory):
    files_data = get_file_stats(directory)

    print("TOP самых больших файлов:")
    for file in show_top_files_by_size(files_data):
        print(file)

    print("\nTOP самых маленьких файлов:")
    for file in show_top_files_by_size(files_data, largest=False):
        print(file)

    print("\nTOP самых длинных имён файлов:")
    for file in show_top_files_by_name_length(files_data):
        print(file)

    print("\nTOP самых коротких имён файлов:")
    for file in show_top_files_by_name_length(files_data, longest=False):
        print(file)

    print("\nTOP самых часто используемых имён файлов:")
    file_names = [file['name'] for file in files_data]
    for name, count, percent in show_top_most_common(file_names):
        print(f"{name}: {count} ({percent:.2f}%)")

    print("\nTOP самых длинных расширений файлов:")
    for file in show_top_files_by_ext_length(files_data):
        print(file)

    print("\nTOP самых коротких расширений файлов:")
    for file in show_top_files_by_ext_length(files_data, longest=False):
        print(file)

    print("\nTOP самых часто используемых расширений файлов:")
    file_exts = [file['ext'] for file in files_data]
    for ext, count, percent in show_top_most_common(file_exts):
        print(f"{ext}: {count} ({percent:.2f}%)")

    print("\nTOP самых длинных полных путей к файлам:")
    sorted_files_by_path_length = sorted(files_data, key=lambda x: x['path_length'], reverse=True)
    for file in sorted_files_by_path_length[:10]:
        print(file)

    print("\nTOP каталогов, где больше всего файлов:")
    for dir, count in show_top_dirs_by_file_count(files_data):
        print(f"{dir}: {count} файлов")

    print("\nTOP каталогов, где больше всего подкаталогов:")
    for dir, count in show_top_dirs_by_subdir_count(directory):
        print(f"{dir}: {count} подкаталогов")

if __name__ == "__main__":
    main("C:/Users/Стас/Desktop/")
```
